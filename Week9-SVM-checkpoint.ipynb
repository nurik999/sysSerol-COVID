#### (6) Vary the complexity of your kernel and evaluate the ability of it to descriminate data both directly and on crossvalidation.

How does the answer of which classification strategy vary based on whether you are performing cross-validation?

```python
# Answer
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
degrees = np.array(range(1,20))*.05
direct = np.zeros(len(degrees))
CV = np.zeros(len(degrees))

kfindindices = np.arange(len(Y))
predictholder = np.zeros([len(Y)])
skf = StratifiedKFold(n_splits=10)

for i in range(1,len(degrees)+1):
    model = SVC(kernel='rbf', gamma=i*.05)
    model.fit(X, Y)
    y = (model.predict(X))
    direct[i-1] = metrics.accuracy_score(Y, y)
    
    for train_index, test_index in skf.split(kfindindices, Y):
        trainY = Y[train_index]
        testY = Y[test_index]
        trainX = X[train_index]
        testX = X[test_index]
    
        model = SVC(kernel='rbf', gamma=i*.05)
        model.fit(trainX, trainY)
        predictholder[test_index] = model.predict(testX)
    CV[i-1] = metrics.accuracy_score(Y, predictholder)
    print(i-1)
plt.plot(degrees, direct, label='Direct Classification')
plt.plot(degrees, CV, label='Cross-validation Prediction')
plt.xlabel("Model Degree")
plt.ylabel('Accuracy')
plt.title('Loadings Plot - PLS-DA')
plt.legend()
```

I would expect that as gamma increases goes up, the direct fitting classification would get better. However, this would eventually lead to overfitting. From this plot, the best gamma value appears to be around 0.45.

#### (7) Your experimental collaborator asks you to determine whether she can collect a fewer number of measurements (cells) and still obtain the same classification efficiency when building a new estimator. How might you determine this? Can she?

Collecting fewer number of measurements means fewer cells. I will delete 6000 rows at random.

```python
import random
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
print('Original CV accuracy: ', CV[8])
c = np.c_[X, Y]

for i in range(6000):
    c = np.delete(c, random.randint(0,len(c))-1, axis=0)
Y1 = c[:, -1]
X1 = c[:,:-1]
model = SVC(kernel='rbf', gamma = 0.45)
model.fit(X1, Y1)
y = (model.predict(X1))
direct = metrics.accuracy_score(Y1, y)

kfindindices = np.arange(len(Y1))
predictholder = np.zeros([len(Y1)])
skf = StratifiedKFold(n_splits=10)

for train_index, test_index in skf.split(kfindindices, Y1):
    trainY = Y1[train_index]
    testY = Y1[test_index]
    trainX = X1[train_index]
    testX = X1[test_index]
    
    model = SVC(kernel='rbf')
    model.fit(trainX, trainY)
    predictholder[test_index] = model.predict(testX)
cv = metrics.accuracy_score(Y1, predictholder)
print('New CV accuracy: ', cv)
```

This increases the model accuracy because there are less overlapping points. It is easier to classify the points when there is less overlapping data. However, adding large amounts of new cells to predict would be more of a problem.

#### (8) What if you can change the numbers of individual cell types that you collect? Is the model you obtain sensitive to the number of cells in each group to the same extent? Try this with one cell population.

```python
# Answer
print('Original CV accuracy: ', CV[8])
c2 = np.c_[X, Y]
for i in range(10000):
    remove = random.randint(0,len(c2)-1)
    if Cell[remove] == "MCF7":
        c2 = np.delete(c2, remove, axis=0)
Y2 = c2[:, -1]
X2 = c2[:,:-1]
model = SVC(kernel='rbf', gamma = 0.45)
model.fit(X2, Y2)
y = (model.predict(X2))
direct = metrics.accuracy_score(Y2, y)

kfindindices = np.arange(len(Y2))
predictholder = np.zeros([len(Y2)])
skf = StratifiedKFold(n_splits=10)

for train_index, test_index in skf.split(kfindindices, Y2):
    trainY = Y2[train_index]
    testY = Y2[test_index]
    trainX = X2[train_index]
    testX = X2[test_index]
    
    model = SVC(kernel='rbf')
    model.fit(trainX, trainY)
    predictholder[test_index] = model.predict(testX)
cv = metrics.accuracy_score(Y2, predictholder)
print('New CV accuracy: ', cv)
```

The accuracy increased. This makes sense as it is easier to classify the points when there are less points overlapping. By having a majority of one cell type, the model will have a higher accuracy as it is able to primarily classify cells as that one type. However, this model will likely not classify new MCF7 cells very well. The stratified K-fold has less MCF7 to predict because these cells have been dropped from the matrix. The accuracy of prediction is increasing for the other two cell types but decreasing for this one.
